# Prueba de pintar los nodos

```python
from model.NetworkGraphPlots import NodeDegreeHeatMapPlot
import matplotlib.pyplot as plt
from model.NodeClass import Node
```

```python
from matplotlib.lines import Line2D

fig, ax = plt.subplots(figsize=(10, 10))

x, y = Node.obtain_x_y_vectors(graph.nodes)
degrees = Node.obtain_degree_vector(graph.nodes)
types = Node.obtain_type_vector(graph.nodes)
names = Node.obtain_name_vector(graph.nodes)

hl3_indices = [i for i, t in enumerate(types) if t == "HL3"]
hl2_1_indices = [i for i, t in enumerate(types) if t == "HL2-1"]


colors = [
    'red' if names[i] in potentially_aggregation_nodes else 'green'
    for i in range(len(names))
]

size = 100

print(potentially_aggregation_nodes)
print(names)
print(colors)



formatting_names = []


ax.scatter(
    [x[i] for i in hl3_indices],
    [y[i] for i in hl3_indices],
    marker='o',
    color=[colors[i] for i in hl3_indices],
    s=size,
)

ax.scatter(
    [x[i] for i in hl2_1_indices],
    [y[i] for i in hl2_1_indices],
    marker='o',
    color=[colors[i] for i in hl2_1_indices],
    s=size,
)

for i in range(len(graph.nodes)):
    ax.text(x[i], y[i], names[i], fontsize=8, color='black', alpha=0.9)


for link in graph.links:
    a = link.a.pos
    b = link.b.pos
    
    if link.a.name in potentially_aggregation_nodes and link.b.name in potentially_aggregation_nodes:
        linewidth = 1
        alpha = 0.5
        color = 'black'
    else:
        linewidth = 1
        alpha = 0.5
        color = 'gray'
    
    ax.plot([a[0], b[0]], [a[1], b[1]], 'k-', lw=linewidth, alpha=alpha, color=color)
    # ax.text(
    #     (a[0] + b[0]) / 2,
    #     (a[1] + b[1]) / 2,
    #     f'{link.distance_km} km',
    #     fontsize=8,
    #     color='black',
    #     alpha=0.7,
    # )

margin = 7 # km
min_x = min(x) - margin
max_x = max(x) + margin
min_y = min(y) - margin
max_y = max(y) + margin
ax.set_xlim(min_x, max_x)
ax.set_ylim(min_y, max_y)

# ax.set_xlabel('X Coordinate (km)')
# ax.set_ylabel('Y Coordinate (km)')
ax.axis('off')

legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='HL2-1', markerfacecolor='red', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='HL3', markerfacecolor='green', markersize=10)
]

ax.legend(handles=legend_elements, loc='upper right')

# ax.legend()
plt.show()
```
